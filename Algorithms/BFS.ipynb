```python
def BFS(graph,start):
    tree = Graph(graph.num_verts())
    visited = []
    queue = []
    queue.append(start)
    v = queue.pop()
    for d in graph[v]:
            if d in visited :
                continue
            else :
                queue.append(d)
            
    visited.append(v)
    return queue
```
