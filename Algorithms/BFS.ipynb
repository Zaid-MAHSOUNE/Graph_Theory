```sage
from collections import deque

def BFS(graph,start):
    visited = []
    queue = deque()

    queue.append(start)
    visited.append(start)

    while queue :
        v = queue.popleft()
        if v not in visited :
            visited.append(v)

        for n in graph.neighbors(v):
            if n not in visited :
                queue.append(n)
                
    return visited
```
