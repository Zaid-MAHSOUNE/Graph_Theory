```python
def DFS(graph,start):
    visited = []
    stack = []

    stack.append(start)
    visited.append(start)

    while stack :
        v = stack.pop()
        if v not in visited :
            visited.append(v)

        for n in graph.neighbors(v):
            if n not in visited :
                stack.append(n)
                        
    return visited
```
